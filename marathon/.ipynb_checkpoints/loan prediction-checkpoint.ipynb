### valeurs manquantes

```python
def missing_value_rate(data):
    return 100 * data.isna().sum() / data.shape[0]
```

```python
missing_value_rate(train_perf)
```

```python
del train_perf["referredby"]
train_perf
```

```python
from datetime import datetime

creationdate = train_perf["creationdate"]
approveddate = train_perf["approveddate"]

date_diff = []


for create, approved in zip(creationdate, approveddate):
    create = create.split(".")[0]
    create = datetime.strptime(create, '%Y-%m-%d %H:%M:%S').timestamp()
    approved = approved.split(".")[0]
    approved = datetime.strptime(approved, '%Y-%m-%d %H:%M:%S').timestamp()
    date_diff.append((approved - create) / 60)
# print(date_diff)
del train_perf["creationdate"]
del train_perf["approveddate"]
train_perf["approveddelai"] = date_diff
train_perf
```

```python
good_bad_flag = train_perf["good_bad_flag"]
good_bad_flag = good_bad_flag.replace("Good", 1)
good_bad_flag = good_bad_flag.replace("Bad", 0)
good_bad_flag
```
